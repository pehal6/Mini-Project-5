Compare the performance(time taken) of C++, R and python to convert 200MB, 400MB, 600MB, 800MB  and 1000MB text files to upper case.

```python
import matplotlib.pyplot as plt

# File sizes (in MB)
file_sizes = [200, 400, 600, 800, 1000]

# Time taken for each language (in seconds)
cpp_time = [15, 25, 36, 50, 60]
r_time = [20, 35, 45, 60, 80]
python_time = [25, 40, 53, 75, 100]

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(file_sizes, cpp_time, 'o-r', label='C++')
plt.plot(file_sizes, r_time, 's-m', label='R')
plt.plot(file_sizes, python_time, '^-c', label='Python')

plt.title("Performance Comparison: C++, R, Python")
plt.xlabel("File Size (MB)")
plt.ylabel("Time Taken (Seconds)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
